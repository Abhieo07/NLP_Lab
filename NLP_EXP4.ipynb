```python
import pdfplumber
import spacy
import re
import PyPDF2
import nltk
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
def extract_text_from_pdf(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text()
    return text
```

```python
# Function to extract the name 
def extract_name(text):
    doc = nlp(text)
    for ent in doc.ents:
        if ent.label_ == "PERSON":
            return ent.text
    return "Name not found"
```

```python
# Function to extract technical skills using regex (can be customized for more specific skills)
def extract_skills(text):
    # Commonly used technical skills keywords
    skills_keywords = [ 
        "Python", "Java", "C++", "JavaScript", "SQL", "Machine Learning", "Data Analysis",  
        "Django", "Flask", "React", "Node.js", "HTML", "CSS", "NoSQL", "MongoDB",  
        "PostgreSQL", "MySQL", "AWS", "Azure", "Docker", "Kubernetes", "Git", "Linux",  
        "Agile Methodology", "Scrum", "REST APIs", "GraphQL", "TensorFlow", "PyTorch",  
        "Natural Language Processing", "Computer Vision", "Cloud Computing", "Microservices",  
        "Blockchain", "Solidity", "Embedded Systems", "MATLAB", "Simulink", "VHDL",  
        "Verilog", "FPGA", "Assembly Language", "Cybersecurity", "Penetration Testing",  
        "Wireshark", "JIRA", "CI/CD", "Ansible", "Terraform", "Network Protocols",  
        "TCP/IP", "Firewalls", "Load Balancers", "Web Development", "Mobile App Development",  
        "Android", "iOS", "Swift", "Kotlin", "Flutter", "React Native", "Unit Testing",  
        "Integration Testing", "Selenium", "Jenkins", "Hadoop", "Spark", "Apache Kafka",  
        "Data Warehousing", "ETL Processes", "Big Data", "R", "Pandas", "NumPy",  
        "Scikit-learn", "SQLAlchemy", "Redis", "Elasticsearch", "Prometheus", "Grafana",  
        "TensorFlow Lite", "OpenCV", "OpenGL", "CUDA", "Virtualization", "VMware",  
        "Hyper-V", "PLC Programming", "ROS (Robot Operating System)", "3D Printing",  
        "Game Development", "Unity", "Unreal Engine", "WebAssembly", "Shell Scripting",  
        "Bash", "PowerShell", "Jupyter Notebooks", "LaTeX", "ARM Architecture",  
        "Quantum Computing", "IoT (Internet of Things)", "Arduino", "Raspberry Pi",  
        "SCADA Systems", "Wireless Networks", "Signal Processing", "API Development",  
        "Object-Oriented Programming", "Functional Programming"
        
    ]
    
    skills = []
    for skill in skills_keywords:
        if re.search(rf"\b{skill}\b", text, re.IGNORECASE):
            skills.append(skill)
    return skills if skills else "Skills not found"

```

```python
# Function to extract experience (years) using regex
def extract_experience(text):
    experience_pattern = re.compile(r"(\d+)\+?\s*(years|yrs)?\s*experience", re.IGNORECASE)
    match = experience_pattern.search(text)
    if match:
        return match.group(1) + " years"
    else:
        return "Experience not found"
```

```python
# Main function to parse the resume
def parse_resume(pdf_path):
    # Extract text from the PDF
    text = extract_text_from_pdf(pdf_path)
    
    # Extract relevant information
    name = extract_name(text)
    skills = extract_skills(text)
    experience = extract_experience(text)
    
    # Print extracted information
    print("Name:", name)
    print("Skills:", skills)
    print("Experience:", experience)
```

```python
# Example usage
pdf_path = "resume-computer-engineering.pdf" 
parse_resume(pdf_path)
```
